### Selecting Specific rows and columns using iloc and loc

```python
# Using loc and iloc(integers) - used to select or extract specific rows and columns. The first parameter refers to the row and the second refer to the columns(you can select more than one column). iloc selects using index integers while loc selects using labels

dataframe.iloc[[0, 25], [1, 4]]
```

```python
dataframe.loc[0, ['Survived', 'Sex']]
```

```python
dataframe.loc[5:10, "Name":"Ticket"]
```

```python
# performing conditional selection with loc
dataframe.loc[dataframe.Sex=='male']
```

```python
# parsing several conditions
dataframe.loc[(dataframe.Sex=='female') & (dataframe.Age > 35)]
```
